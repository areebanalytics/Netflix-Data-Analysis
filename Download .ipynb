```python
import pandas as pd
import zipfile

with zipfile.ZipFile("/content/netflix_titles.csv.zip", 'r') as z:
    z.extractall("/content/")

df = pd.read_csv("/content/netflix_titles.csv")

df.head()
```

```python
df['director'].fillna("unknown",inplace=True)
df['cast'].fillna("unknown",inplace=True)
df['country'].fillna("unknown",inplace=True)
df['date_added'].fillna("not available",inplace=True)
df['duration'].fillna("not available",inplace=True)
most_common_rating=df['rating'].mode()[0]
df['rating'].fillna(most_common_rating,inplace=True)
df.isnull().sum()
```

```python
df['duration_num']=df['duration'].str.extract('(\d+)')
df['duration_num']=pd.to_numeric(df['duration_num'])
df[['duration','duration_num']].head(10)
```

```python
df['movie_duration']=df.apply(lambda x: x['duration_num'] if 'min' in str(x['duration'])else None,axis=1)
df['tv_seasons']=df.apply(lambda x: x['duration_num']if 'season' in str(x['duration'])else None,axis=1)
df[['duration','movie_duration','tv_seasons']].head(10)
```
